```python
import pandas as pd 
import numpy as np 
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams["figure.figsize"]=(12, 8)
```

#### loading in our data  set
##### movies data set , source : kaggle

```python
df = pd.read_csv(r"C:\Users\ANANY\Desktop\movies.csv")
```

```python
df.head()
```

##### checking for null values

```python
df.isnull().isnull().sum()
```

##### no null values present in the dataset. 

#### Data Cleaning

```python
df.dtypes
```

```python
df['budget'] = df['budget'].fillna(0)
df['gross'] = df['votes'].fillna(0)
```

##### removing decimal for avoiding errors .

```python
df["budget"] = df["budget"].astype("int64")
df["gross"] = df["gross"].astype("int64")
```

```python
df.head()
```

```python
df.sort_values(by=["gross"], inplace=False, ascending=False).head()
```

##### drop any duplicates

```python
df.drop_duplicates().head()
```

```python
df["score"].fillna(0)
df["rating"].fillna("R")
```

##### Checking for correlation using scatterplot

```python
plt.scatter(x=df["budget"], y=df["gross"])
plt.title("Budget vs gross earnings")
plt.xlabel("gross earning")
plt.ylabel("budget for films")

plt.show()
```

##### plotting budget vs gross using seaborn 

```python
sns.regplot(x="budget", y="gross",data=df,scatter_kws={"color":"black"}, line_kws={"color":"purple"})
plt.show()
```

```python
df.corr(method="spearman")
```

##### positive correlation between gross and budget

```python
correlation_matrix = df.corr(method="spearman")
```

```python
sns.heatmap(correlation_matrix, annot=True)
plt.title("correlation matrix for numeric features")
plt.show()
```

#### looking at a firm/company correlation with film earning

```python
df_num = df

for col_name in df_num.columns:
    if(df_num[col_name].dtype == "object"):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes

df_num.head()
```

```python
correlation_matrix = df_num.corr(method="spearman")

sns.heatmap(correlation_matrix, annot=True)
plt.title("correlation matrix for numeric features")
plt.show()
```

```python
df_num_corr_matrix = df_num.corr(method="spearman")

corr_pairs = df_num_corr_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
```

```python
high_corr = sorted_pairs[(sorted_pairs) >0.5]
high_corr
```

##### votes and budgets have high correlation to gross earnings of movie.
